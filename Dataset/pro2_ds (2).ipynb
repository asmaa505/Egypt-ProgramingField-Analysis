```python
# deal with missing values

import numpy as np
import pandas as pd

def clean_exp(x):
    try:
        if isinstance(x, str):
            x = x.lower().strip()

            # months
            if "month" in x:
                return float(x.split()[0]) / 12

            # range
            elif "-" in x:
                parts = x.split("-")
                return (float(parts[0].strip()) + float(parts[1].split()[0].strip())) / 2

            # years
            else:
                return float(x.split()[0])

        return x

    except:
        return np.nan


col = "years of experiences"

df[col] = df[col].apply(clean_exp)

df[col] = pd.to_numeric(df[col], errors="coerce")

df[col] = df[col].fillna(df[col].median())
```

## Outliers Detection (IQR Method)

This code is used to clean the data, visualize distributions, and detect outliers using the IQR method.

- Final Output

All detected outliers are combined into one dataframe
If found → displayed with total count
If not → message: "No outliers found"

- Result:

This helps identify abnormal values in the dataset based on statistical distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# =========================
# 1) CLEAN TITLE COLUMN
# =========================
df["title"] = df["title"].str.lower().str.strip().str.replace(r"\s+", " ", regex=True)

# =========================
# 2) NUMERICAL COLUMNS
# =========================
num_cols = df.select_dtypes(include=['number']).columns

# =========================
# 3) BOX PLOTS (BEFORE REMOVING ANYTHING)
# =========================
for col in num_cols:
    plt.figure(figsize=(6,4))
    df.boxplot(column=col)
    plt.title(f"Boxplot of {col}")
    plt.show()

for col in num_cols:
    plt.figure(figsize=(12,5))
    sns.boxplot(x="title", y=col, data=df)
    plt.xticks(rotation=90)
    plt.title(f"{col} by Title")
    plt.show()

# =========================
# 4) OUTLIERS DETECTION (IQR METHOD)
# =========================
all_outliers = []

for title, group in df.groupby("title"):
    for col in num_cols:
        Q1 = group[col].quantile(0.25)
        Q3 = group[col].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = group[(group[col] < lower) | (group[col] > upper)].copy()

        if not outliers.empty:
            outliers["outlier_column"] = col
            outliers["title_group"] = title
            all_outliers.append(outliers)

# =========================
# 5) FINAL OUTLIERS TABLE
# =========================
if all_outliers:
    outliers_df = pd.concat(all_outliers, ignore_index=True)
    print("TOTAL OUTLIERS FOUND:", len(outliers_df))
    display(outliers_df)
else:
    print("No outliers found")
```

###  Handling Outliers using IQR Clipping

In this step, outliers were treated using the **IQR (Interquartile Range)** method with clipping instead of removal.

- For each numerical column:
  - Calculated Q1 (25th percentile) and Q3 (75th percentile)
  - Computed IQR = Q3 - Q1
  - Defined lower and upper bounds:
    - Lower = Q1 - 1.5 * IQR
    - Upper = Q3 + 1.5 * IQR

- Values outside this range were **clipped** to the nearest boundary.

 This approach preserves all data points while reducing the impact of extreme values.

 Dataset shape remains unchanged before and after clipping.

```python
import pandas as pd


df_fixed = df.copy()



num_cols = df_fixed.select_dtypes(include=['number']).columns


for col in num_cols:
    Q1 = df_fixed[col].quantile(0.25)
    Q3 = df_fixed[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df_fixed[col] = df_fixed[col].clip(lower, upper)


print("Before shape:", df.shape)
print("After fixing outliers:", df_fixed.shape)
```

## Salary vs Experience Correlation (Per Job Title)

This analysis measures how salary changes with years of experience for the top 10 job titles, then visualizes the results using a heatmap.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df["salary"] = pd.to_numeric(df["salary"], errors="coerce")
df["years of experiences"] = pd.to_numeric(df["years of experiences"], errors="coerce")
df_clean = df.dropna(subset=["salary", "years of experiences", "title"])

top_jobs = df_clean["title"].value_counts().nlargest(10).index
df_top = df_clean[df_clean["title"].isin(top_jobs)]


corr_results = {}

for job in top_jobs:
    subset = df_top[df_top["title"] == job]
    
    if len(subset) > 1:  
        corr = subset["salary"].corr(subset["years of experiences"])
        corr_results[job] = corr

corr_df = pd.DataFrame.from_dict(corr_results, orient="index", columns=["Correlation"])

# =========================
# 4. Heatmap
# =========================
plt.figure(figsize=(6,5))
sns.heatmap(corr_df, annot=True, cmap="coolwarm", center=0, fmt=".2f")

plt.title("Salary vs Experience Correlation per Job")
plt.ylabel("Job Title")
plt.show()
```

```python
from scipy.stats import f_oneway

roles = ["backend developer", "data scientist", "data engineer", ".net developer"]

groups = []

for role in roles:
    group = df_fixed[df_fixed["title"] == role]["salary"].dropna()
    groups.append(group)

# ANOVA
f_stat, p_value = f_oneway(*groups)

print("F-statistic:", f_stat)
print("P-value:", p_value)
```

## ANOVA After Fixing Experience Mixing Issue

After the previous ANOVA test, we suspected that the results might be misleading because we were mixing different experience levels together.  
So, we repeated the test by separating the data based on **experience level**.

#### Result

“No statistically significant difference was found between job roles within each experience level, possibly due to overlapping salary distributions and limited sample sizes in some groups.”

```python
from scipy.stats import f_oneway

experiences = ["junior", "mid", "senior"]

for exp in experiences:
    print(f"\n=== {exp.upper()} LEVEL ===")
    
    subset = df_fixed[df_fixed["experience level"] == exp]

    roles = ["backend developer", "data scientist", "data engineer", ".net developer"]
    
    groups = []
    
    for role in roles:
        group = subset[subset["title"] == role]["salary"].dropna()
        
        if len(group) > 1:
            groups.append(group)

  
    if len(groups) >= 2:
        f_stat, p_value = f_oneway(*groups)

        print("F-statistic:", f_stat)
        print("P-value:", p_value)

        if p_value < 0.05:
            print(" Significant difference between roles")
        else:
            print(" No significant difference between roles")
    else:
        print("Not enough data for ANOVA")
```

## Cohen’s d (Effect Size Analysis)

After performing ANOVA, which tests whether there is a statistically significant difference between job roles, we applied **Cohen’s d** to measure the **strength (magnitude) of the difference** between two groups.

Unlike ANOVA, which only tells us whether a difference exists or not, Cohen’s d helps us understand **how large or small that difference actually is in practice**.

---

###  Result:
- Cohen’s d = **-0.3057**

---

##  Interpretation
- The value is close to 0  
- This indicates a **small effect size**  
- The negative sign only shows the direction of the difference (which group has a slightly higher mean)

---

##  Conclusion
Although ANOVA suggested no statistically significant difference between roles, Cohen’s d confirms that even if a difference exists, it is **small and not practically meaningful** in this dataset.

```python

import numpy as np

def cohens_d(x, y):
    return (np.mean(x) - np.mean(y)) / np.sqrt((np.std(x)**2 + np.std(y)**2) / 2)

ds = df_fixed[df_fixed["title"]=="data scientist"]["salary"]
be = df_fixed[df_fixed["title"]=="backend developer"]["salary"]

print(cohens_d(ds, be))
```

## Confidence Intervals for Salary by Job Title
### Why we did this

After comparing salaries using ANOVA and Cohen’s d, we wanted a more interpretable statistical range instead of just averages.
So we used Confidence Intervals to understand the uncertainty around each job’s average salary.

### What the results show

Each job title has a mean salary and a range (CI lower → CI upper)
The CI shows where the true average salary is likely to fall with 95% confidence

### Key Insights
1. Highest paying roles

Backend Developer, Java Developer, and Frontend Developer are among the highest average salaries
They also have relatively wide ranges, meaning salary variability is high

2. Tech vs non-tech gap

Technical roles (Software, Backend, ML, Data Science) generally have higher salaries
Non-technical roles (Accountant, Pharmacist, Graphic Designer) are at the lower end

3. Salary uncertainty
Some roles like Machine Learning Engineer and Data Scientist have wide confidence intervals

 This means salaries are not very stable and vary a lot

4. Data quality insight
Roles with fewer records show wider confidence intervals

 meaning less reliable estimates

### Final Conclusion

Confidence intervals show that while there are clear differences in average salaries between roles, there is also a high level of variation within each role, especially in tech-related jobs.

This helps us understand not just the salary level, but also how stable or variable each job’s salary is in the market.

```python
import scipy.stats as st

ci_results = []

for role in df_fixed["title"].unique():
    data = df_fixed[df_fixed["title"] == role]["salary"].dropna()
    
    if len(data) < 5:
        continue

    mean = data.mean()
    ci = st.t.interval(0.95, len(data)-1, loc=mean, scale=st.sem(data))

    ci_results.append({
        "title": role,
        "mean": mean,
        "ci_lower": ci[0],
        "ci_upper": ci[1]
    })

ci_df = pd.DataFrame(ci_results)
ci_df.sort_values("mean", ascending=False)
```
