```python
# Integration
import pandas as pd

df1 = pd.read_csv("Web-Developers-Salaries-Egypt-2024.csv")
df2 = pd.read_csv("wuzzuf_jobs.csv")


df2.rename(columns={
    "Years of Experiences ": "Years of Experiences"
}, inplace=True)


df2.drop(columns=["Unnamed: 0", "TimeStamp"], inplace=True)


df = pd.concat([df1, df2], ignore_index=True)
```

```python
df["Salary"] = pd.to_numeric(df["Salary"], errors='coerce')
df["Years of Experiences"] = df["Years of Experiences"].apply(clean_exp)
df["Work Hour"] = pd.to_numeric(df["Work Hour"], errors='coerce')

print(df.dtypes)
```

```python
# deal with missing values

import numpy as np

def clean_exp(x):
    try:
        if isinstance(x, str):
            x = x.lower().strip()

            # month
            if "month" in x:
                return float(x.split()[0]) / 12

            # range
            elif "-" in x:
                parts = x.split("-")
                return (float(parts[0]) + float(parts[1].split()[0])) / 2

            # years
            else:
                return float(x.split()[0])

        return x

    except:
        return np.nan  
    

df["Years of Experiences"] = df["Years of Experiences"].apply(clean_exp)


df["Years of Experiences"] = pd.to_numeric(df["Years of Experiences"], errors="coerce")

df["Years of Experiences"].fillna(df["Years of Experiences"].median(), inplace=True)

#-----------------------------------------------------------------------------------------------------#

```

```python
df["City of Company site"].fillna(df["City of Company site"].mode()[0], inplace=True)
```

```python
# Date of Salary colunm
print(f'count of not null values: {df["Date of Salary"].count()}\ncount of null values: {df["Date of Salary"].isnull().sum()}')
```

```python
df["Date of Salary"] = pd.to_datetime(df["Date of Salary"], errors="coerce")
df["Date of Salary"] = df["Date of Salary"].ffill().bfill()
```

```python
num_cols = df.select_dtypes(include=['number']).columns

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\nColumn: {col}")
    print(f"Number of outliers: {outliers.shape[0]}")
```
